```python
deaths_df = pd.read_csv('time_series_covid_19_deaths.csv')
```

```python
deaths_df[deaths_df['Country/Region'] == 'Brazil']
```

```python
columns = deaths_df.keys()
columns
```

```python
deaths_df['5/29/21'].sum()
```

```python
deaths_df = deaths_df.loc[:, columns[4]:columns[-1]]
```

```python
dates = deaths_df.keys()
y = []
for i in dates:
  y.append(deaths_df[i].sum())
```

```python
len(y)
```

```python
x = np.arange(len(dates)).reshape(-1, 1)
```

```python
x.shape
```

```python
x # x  siginifica as datas 1 = 1 dia e assim por diante
```

```python
forecast = np.arange(len(dates) + 10).reshape(-1, 1)
```

```python
x_treinamento, x_test, y_treinamento, y_test = train_test_split(x, y, test_size = 0.15, shuffle = False)
```

```python
print('testes x = ', len(x_test))
print('testes y = ', len(y_test))
print('treinament x = ', len(x_treinamento))
print('treinamento y = ', len(y_treinamento))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_treinamento)
x_test_poly = poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train_poly, y_treinamento)
```

```python
poly_test_pred = regressor.predict(x_test_poly)
poly_test_pred
```

```python
plt.plot(poly_test_pred, linestyle = 'dashed') #previsoes
plt.plot(y_test) #real (laranja)
plt.legend(['Predictions', 'Death cases']);
```

```python
print('MAE:', mean_absolute_error(poly_test_pred, y_test))
print('MSE:', mean_squared_error(poly_test_pred, y_test))
print('RMSE:', np.sqrt(mean_absolute_error(poly_test_pred, y_test)))
```

```python
x_treinamento_all = poly.transform(forecast)
```

```python
poly_pred_all = regressor.predict(x_treinamento_all)
len(poly_pred_all)
```

```python
plt.plot(forecast[:-10], y, color='red')
plt.plot(forecast, poly_pred_all, linestyle='dashed')
plt.title('Deaths of COVID-19')
plt.xlabel('Days since 1/22/2020')
plt.ylabel('Number of deaths')
plt.legend(['Death cases', 'Predictions']);
```
